```python
import re
import jieba.analyse
```

### 对中文数据集进行预处理

```python
def preprocess(input_file, output_file):
    reader = open(input_file, 'r',encoding='utf-8')
    writer = open(output_file, 'w',encoding='utf-8')
    line=reader.readline()
    while line:
        content = line.split()[1:]
        content = re.sub(r'\W+', ' ',str(content), flags=re.U)
        writer.write(str(content))
        writer.write('\n')
        line = reader.readline()
    reader.close()
    writer.close()
```

### 实现分词，并且提取关键词

```python
def segment(filename):
    comment = open(filename, 'r', encoding='utf-8')
    comment = comment.read()
    # cutword = jieba.cut(comment)
    jieba.analyse.set_stop_words('./data/stopwords.txt')
    global key_words
    key_words = jieba.analyse.extract_tags(comment, topK=2000, withWeight=True,
                                           allowPOS=('nb', 'n', 'nr', 'ns', 'a', 'ad', 'an', 'nt', 'nz', 'v', 'd'))
    for word, fre in key_words:
        print(word, fre)
```

### 将segment函数中得到的关键词以词云的形式展示

```python
def word_cloud():
    from wordcloud import WordCloud
    import matplotlib.pyplot as plt
    def show_img(wc):
        plt.figure()
        plt.imshow(wc)
        plt.axis("off")
    # 实例化，通过font_path传入一个支持中文的字体
    wc = WordCloud(font_path="./data/simhei.ttf",
                   max_words=2000,
                   width=1920,
                   height=1080,
                   background_color="black",
                   margin=5)
    # 传入[（key，weight）,...]列表生成词云
    wc.generate_from_frequencies(dict(key_words))
    # 保存图片到本地
    wc.to_file('santi2_textrank_filtered(nb_n_nr_ns_a_ad_an_nt_nz_v_d).png')
    show_img(wc)
if __name__ == '__main__':
    preprocess('./data/cnews.train.txt', './data/cnews.train_preprocess.txt')
    segment('./data/cnews.train_preprocess.txt')
    word_cloud()
```
